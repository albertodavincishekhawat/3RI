```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
data = np.random.randint(40, 99, 35)
data.sort()
```

```python
# Line Plot
plt.title("Line Plot")
plt.xlabel("Numbers")
plt.ylabel("Marks")
sns.lineplot(data, palette="mako")
plt.show()
```

```python
df = pd.read_excel("files/tips.xlsx")
```

```python
df.head()
```

```python
# Count Plot
plt.figure(figsize=(4,3))
plt.title("Count Plot")
plt.xlabel("Gender")
plt.ylabel("Count")
sns.countplot(df, x="gender", palette="Reds")
plt.show()
```

```python
plt.title("Count Plot")
plt.xlabel("Gender")
plt.ylabel("Count")
sns.countplot(df, x="gender", palette="Reds", hue="smoker")
plt.show()
```

```python
# KDE plot  
plt.title("KDE Plot")
sns.kdeplot(df, x="total_bill", fill=True, hue="time", palette="rocket")
plt.show()
```

```python
# Hist plot
plt.title("Histogram")
sns.histplot(df, x="total_bill", hue="time", kde=True)
plt.show()
```

```python
# KDE plot
plt.title("Histogram")
sns.kdeplot(df, x="total_bill", hue="day", fill=True)
plt.show()
```

```python
sns.jointplot(df, x="total_bill", y="tip", palette="mako", hue="gender")
plt.show()
```

```python
iris = sns.load_dataset("iris")
```

```python
iris["species"].value_counts()
```

```python
# Joint Plot
sns.jointplot(iris, x="petal_length", y="petal_width", hue="species")
plt.show()
```

```python
# Pair Plot
sns.pairplot(iris, hue="species")
plt.show()
```

```python
# Pair Plot
sns.pairplot(df, hue="gender")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.violinplot(df, x="tip", hue="day", palette="mako")
plt.show()
```
